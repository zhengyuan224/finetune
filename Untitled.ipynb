```python
import pandas as pd
```

```python
df = pd.read_json("round1_train_data.jsonl",lines=True)
```

```python
df
```

```python
answer_list=[]
for _,content in df.iterrows():
    for i in content['questions']:
        opt_list=[]
        k = 0
        for j in i['options']:
        
            opt_list.append(chr(ord('A')+k) + ":" + j)
            k = k + 1
        
        answer_list.append({"problem":content['problem']+"\n"+i['question'],"answer":i['answer'],"opt":"\n".join(opt_list)})
```

```python
df = pd.DataFrame(answer_list)
```

```python
df['instruction'] = df['problem'] + "\n"+ df['opt']
```

```python
df['output'] = df['answer']
```

```python
df
```

```python
df['input']= ''
```

```python
df
```

```python
df[['instruction','input','output']].to_json("data.json",orient='records',force_ascii=False)
```
