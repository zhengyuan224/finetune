```python
import pandas as pd
df = pd.read_csv("test.csv")
```

```python
df
```

```python
import requests
from tqdm.notebook import tqdm
url = 'http://localhost:8000/v1/chat/completions'

headers = {

    'Content-Type': 'application/json',
}
```

```python
data = {
"model": "string",
"messages": [
    {
        "role": "user",
        "content": df.iloc[0]['instruction']
    }
],
"n": 1,
"max_tokens":10,
"stream": False,
"do_sample":False
}
response = requests.post(url, headers=headers, json=data)
response.text
```

```python
answer_list=[]
from tqdm.notebook import tqdm
for _,content in tqdm(df.iterrows()):
    data = {
    "model": "string",
    "messages": [
        {
            "role": "user",
            "content": content['instruction']
        }
    ],
    "n": 1,
    "max_tokens":10,
    "stream": False,
    "do_sample":False
    }
    response = requests.post(url, headers=headers, json=data)
    answer_list.append(response.json()['choices'][0]['message']['content'])
```

```python
df['pred_qwen'] = answer_list
```

```python
df.to_csv("pred_qwen.csv",index=False)
```
